```python
import numpy as np
```

```python
# fancy indexing using array
arr = np.arange(20,31)
arr_res = np.array([3,4,6])
print("elements at index 3,4,5:")
print(arr[arr_res])
```

```python
# converting 1D to 2D array using fancy indexing
arr = np.arange(1,10)
arr_indices = np.array([[5,3],[4,5]])
arr_2d = arr[arr_indices]
print(arr_2d)
```

```python
arr = np.arange(27).reshape(3,3,3)
arr_dept = np.array([0,1])
arr_row = np.array([1,2])
arr_col = np.array([0,2])
arr_3d = arr[arr_dept,arr_row,arr_col]
print("arr_dept[0,1], arr_row[1,2], arr_col[0,2]:")
print(arr_3d)
```
